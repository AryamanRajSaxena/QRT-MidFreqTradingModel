```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import joblib

# 1. Load Data
# Assuming you've already downloaded the parquet files
features_df = pd.read_parquet('features.parquet') # MultiIndex: (Feature, Stock)
returns_df = pd.read_parquet('returns.parquet')   # Index: Date, Columns: Stocks
universe_df = pd.read_parquet('universe.parquet') # Index: Date, Columns: Stocks

# 2. Align Features and Returns (The "Shift")
# We use today's features to predict tomorrow's returns.
target = returns_df.shift(-1)

```

```python
# Verify the index type
print(type(features_df.index))

# If it says 'Index' (strings) instead of 'DatetimeIndex', convert it:
features_df.index = pd.to_datetime(features_df.index)
returns_df.index = pd.to_datetime(returns_df.index)
```

```python
import pandas as pd

# 1. Flatten Features (X) and shifted Returns (y)
# Assuming 'features_df' is your wide-format features and 'returns_df' is wide returns
X_stacked = features_df.loc['2005':'2019'].stack(level=1)
y_stacked = returns_df.shift(-1).loc['2005':'2019'].stack()

# 2. Align them using a temporary DataFrame
# This 'inner' join ensures we only keep (Date, Stock) pairs that exist in both
train_data = pd.DataFrame({'target': y_stacked}).join(X_stacked, how='inner')

# 3. Apply the Universe Filter
# We only want rows where the universe == 1
universe_stacked = universe_df.loc['2005':'2019'].stack()
train_data = train_data.loc[universe_stacked[universe_stacked == 1].index]

# 4. Drop the "Dead Zone" (The last day of 2019 where target is NaN)
train_data = train_data.dropna(subset=['target'])

# 5. Handle missing feature values (NaNs in RSI, MACD, etc.)
train_data = train_data.fillna(0)

# 6. Final Split into X and y
y_train = train_data['target']
X_train = train_data.drop(columns=['target'])

print(f"Sync complete! X rows: {len(X_train)}, y rows: {len(y_train)}")
```

```python
# 4. Train the Random Forest
# We limit depth to 5 to prevent overfitting (very common in noise-heavy finance data)
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, max_depth=5, n_jobs=-1, random_state=42)
model.fit(X_train, y_train)

# Save the model to use in the function
joblib.dump(model, 'rf_model.pkl')
```
